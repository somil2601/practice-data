```python
import matplotlib.pyplot as plt
import numpy as np

M = 1
m1 = 0.5
m2 = 0.5
D=1
t0 = 500
t = np.linspace(-t0,0, 1000)
eta = (m1*m2)/(m1+m2)**2
tau = (eta/5*M)*(t0-t)
phase = -(1/eta)*(tau**(5/8))

#Second Assumption
M2 = 2
m1_2 = 1
m2_2 = 1
D2 = 3
t0_2 = 500
t2 = np.linspace(-t0_2,0, 1000)
eta2 = (m1_2 * m2_2) / (m1_2 + m2_2)**2
tau2 = (eta2 / 5 * M2) * (t0_2 - t2)
phase2 = -(1 / eta2) * (tau2**(5 / 8))

def derivative(phase):
    return np.gradient(phase)

def derivative(phase2):
    return np.gradient(phase2)


plt.plot(t,derivative(phase), label = 'phase')
plt.plot(t2,derivative(phase2), label = 'phase2')
plt.xlabel('time')
plt.ylabel('Phase [omega(t) = d(phase)/dt]')
plt.title('Phase')
plt.legend()
plt.grid()
plt.show()






```

```python
A = (4 * eta * M * ((2 * M * (derivative(phase))**(2/3)))) / (D)
A1 = (4 * eta * M2 * ((2 * M2 * (derivative(phase2))**(2/3)))) / (D2)
plt.plot(t, A, label = 'Amplitude')
plt.plot(t2, A1, label = 'Amplitude2')
plt.xlabel('time')
plt.ylabel('Amplitude')
plt.title('Amplitude')
plt.legend()
plt.grid()
plt.show()
```

```python
signal = A * np.cos(2*phase)
signal2 = A1 * np.cos(2*phase2)

plt.plot(t, signal, label = 'Signal [M =1, D=1]')
plt.plot(t2, signal2, label = 'Signal2 [M = 2, D = 3]')
plt.xlabel('time')
plt.ylabel('Signal')
plt.title('Signal')
plt.legend()
plt.grid()
plt.show()
```
